## *Muestreo.*

T√©cnica para representar  a una poblaci√≥n.

### *Muestreo Aleatorio Probabil√≠stico.*

Cada elemento debe tener la misma probabilidad de ser seleccionado para la muestra.

**-M√©todos de Estratificaci√≥n-**

- **Aleatorio Sistem√°tico.** Tener un sistema aleatorio para la elecci√≥n de los datos
- **Aleatorio Simple**. n√∫meros al azar utilizando m√©todos de programaci√≥n para simular la aleatoriedad.
- **Estratificaci√≥n**. Dividir la Poblaci√≥n en conjunto a segmentos (Estratos) seleccionados proporcionalmente.
- **Conglomerado**. Dividir la poblaci√≥n en grupos, y se seleccionan varios grupos para realizar la muestra.

### *Muestreo No Probabil√≠stico.*

Seleccionado seg√∫n los criterios no aleatorios realizados seg√∫n el experimentador.

**-M√©todos de Estratificaci√≥n-**

- **Por Conveniencia.** Tomar datos convenientes para realizar la muestra.
- **Por juicio.** Basados por la experiencia de un senior.
- **Por Cuotas.** Cuotas con proporciones especificas.
- **Snow Ball.** El primer datos seleccionado elige los siguientes.

# Muestreo - Ejemplo de clase
### Creacion de una muestra aleatoria simple de 50 registros.


```
sample_size = 50
sample = data.sample(n= sample_size,random_state=77)
```

### Revision de la Muestra.

```
sample_size= len(sample)
print (sample_size)
```

# Seleccion sistematica de la  muestra.



####Definicion del intervalo K (tama√±o de la poblacion / tama√±o de la muestra)

```
sample_size = 100
interval_k = population_size//sample_size
```

###Seleccion de Muestra sistematica

```
systematic_sample = data.iloc[::interval_k]
```

```
print ("Muestra obtenida por muestreo sistematico:")
display(systematic_sample)
```
